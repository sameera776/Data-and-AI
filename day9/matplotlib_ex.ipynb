```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("student.csv")
print("Original Data:\n", df, "\n")

# 1 Find missing values
print("Missing values:\n", df.isnull(), "\n")

# 2 Fill missing Age and Marks using mean
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Marks"].fillna(df["Marks"].mean(), inplace=True)

# 3 Clean City column (capitalize properly)
df["City"] = df["City"].str.strip().str.title()

print("Cleaned Data:\n", df, "\n")

#  Analysis
# 1 Average marks
avg_marks = df["Marks"].mean()
print("Average Marks:", avg_marks)

# 2 Top scoring student
top_student = df.loc[df["Marks"].idxmax()]
print("\nTop Scoring Student:\n", top_student)

# 3 Students scoring below 70
below_70 = df[df["Marks"] < 70]
print("\nStudents scoring below 70:\n", below_70)

#  Visualization
plt.bar(df["Name"], df["Marks"])
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Marks vs Students")
plt.show()
```
